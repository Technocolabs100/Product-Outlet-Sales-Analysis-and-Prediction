```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
```

```python
Train_data_NoOutletSize = Train_data.drop('Outlet_Size', axis=1)
```

```python
Train_data_NoOutletSize
```

```python
Train_data_ItemWeight_ItemMRP = Train_data_NoOutletSize[['Item_Weight','Item_MRP']]
Train_data_ItemWeight_ItemMRP
```

```python
Train_data_imputed = imputer.fit(Train_data_ItemWeight_ItemMRP)
Train_data_imputed
```

```python
imputer.statistics_
```

```python
X = imputer.transform(Train_data_ItemWeight_ItemMRP)
```

```python
Train_data_tr = pd.DataFrame(X, columns=Train_data_ItemWeight_ItemMRP.columns)
Train_data_tr
```

```python
Train_data_tr.info()
```

```python
Train_data_numerical = Train_data_NoOutletSize[['Item_Visibility','Item_Outlet_Sales','Outlet_Establishment_Year']]
```

```python
Train_data_num = pd.concat([Train_data_tr,Train_data_numerical], axis=1)
```

```python
Train_data_num
```

```python
Train_data_Categorical = Train_data_NoOutletSize[['Item_Fat_Content','Item_Type','Outlet_Location_Type','Outlet_Type']]
```

```python
Train_data_Categorical
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse=False, drop='first')
```

```python
Train_data_Categorical_1hot = encoder.fit_transform(Train_data_Categorical)
Train_data_Categorical_1hot
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
Train_data_num_NoYear_NoTarget = Train_data_num.drop(['Item_Outlet_Sales','Outlet_Establishment_Year'], axis=1)
Train_data_num_NoYear_NoTarget
```

```python
Train_data_num_NoYear_NoTarget_scaled = scaler.fit_transform(Train_data_num_NoYear_NoTarget)
Train_data_num_NoYear_NoTarget_scaled
```

```python
Train_data_num_NoYear_NoTarget_scaled.shape
```

```python
import numpy as np
```

```python
Final_data = np.concatenate((Train_data_num_NoYear_NoTarget_scaled, Train_data_Categorical_1hot), axis=1)
```
